```python
def addstudent():
    student_name=input("enter the student name")
    idno=input("enter student id")
    gpa=float(input("enter student GPA"))
    print("student  details    :")
    print("student name:",student_name)
    print("student id number",idno)
    print("student GPA",gpa)
    mydict={}
    mydict["name"]=student_name
    mydict["id"]=idno
    mydict["gpa"]=gpa
    students.append(mydict)
    print("the student details was added")
    print("----------------------------------")
def removestudent():
    name=input("enter the student name")
    id=input("enter the student id no")
    
    for i in students:
        if( name in i["name"] and id in i["id"]):
            students.remove(i)
            print("the student details was removed")
            break
    else:
        print("student not found")
    print(students)
    print("-------------------------------------")
def updatestudent():
    name=input("enter the student name")
    id=input("enter the b student id number")
    for i in students:
        if(name in i["name"] and id in i["id"]):
            ch=input("you update  id or gpa press y  add age press x")
            if(ch=='y'):
                id=input("enter update id")
                gpa=input("enter update gpa")
                i["id"]=id
                i["gpa"]=gpa
                print("after update the student details")
                print(i)
            elif(ch=='x'):
                age=int(input("enter  the student age"))
                i['age']=age
        
            print("after update  student detail" )
            print(i)
            break
    else:
        print("student not found")
    print("------------------------------------------")
            
        
def displaystudent():
    name=input("enter  student the name")
    id=input("enter  student id no number")
    for i in students:
        if name in i["name"]  and  id in i["id"] :
            print(i)
            break
    else:
         print("no student found")
    print("------------------------------------")
def searchstudent():
    name=input("enter the name")
    for i in students:
        if(name in i["name"]):
               print(i) 
               break
    else:
         print("student not found")
    print("---------------------------------------")
students=[]
while True:
    print("********************************")
    print("     1.add student")
    print("     2.remove student")
    print("     3.update student")
    print("     4.display student  details")
    print("     5.search student")
    print("     6.exit")
    print("*********************************")
    print("enter the number 1 to 6 your choice")
    num=int(input("enter a number your"))
    if(num==1):
        addstudent()
    elif(num==2):
        removestudent()
    elif(num==3):
        updatestudent()
    elif(num==4):
        displaystudent()
    elif(num==5):
        searchstudent()
    elif(num==6):
        break
    else:
        print("enter between number 1 to 6")
```
